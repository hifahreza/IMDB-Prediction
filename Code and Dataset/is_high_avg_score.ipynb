```python
import pandas as pd
```

```python
df = pd.read_csv('films_fall_2021_training_data - films_fall_2021.csv')
```

```python
df.head()
```

```python
def get_movies(df, col_name):
    # Items can be actors, directors, producers, etc.
    items = df[col_name].unique()
    item_movies = {}
    for item in items:
        item_df = df.loc[df[col_name]==item]
        item_movies_list = list(item_df['imdb_id'].unique())
        item_movies.update({item: item_movies_list})
    return item_movies
```

```python
actor_1_movies = get_movies(df, 'main_actor1_name')
actor_2_movies = get_movies(df, 'main_actor2_name')
actor_3_movies = get_movies(df, 'main_actor3_name')
director_movies = get_movies(df, 'main_director_name')
producer_movies = get_movies(df, 'main_producer_name')
production_co_movies = get_movies(df, 'main_production_company')
```

```python
results = []
col = 'main_actor1_name'
for index, row in df.iterrows():
    actor = row[col]
    imdb_id = row['imdb_id']
#     print(actor, imdb_id)
    movies_included_in_avg = [x for x in actor_1_movies[actor] if x != imdb_id]
#     print(movies_included_in_avg)
    # calculate the average imdb score for this actor for movies_included_in_avg
    _df = df.loc[(df[col]==actor) & (df['imdb_id'].isin(movies_included_in_avg))]
    avg_score = _df['imdb_score'].mean()
    if avg_score >= 6.7:
        dummy_val = 1
    else:
        dummy_val = 0
    result = {
        col: actor,
        'imdb_id': imdb_id,
        f'avg_score_excluding_imdb_id_{col}': avg_score,
        f'is_high_avg_score_{col}': dummy_val
    }
    results.append(result)
actor_1_df = pd.DataFrame(results).set_index(['imdb_id', col])
actor_1_df.head()
```
